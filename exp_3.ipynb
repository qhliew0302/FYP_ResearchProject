### Preprocess and Load Data

```python
import cv2
import os
import numpy as np
```

```python
# Define image size
img_size = (224, 224)

# Define function to load and preprocess images
def load_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, img_size)
    img = img / 255.0
    return img

# Define function to load and preprocess ground truth saliency maps
def load_saliency_map(saliency_path):
    saliency = cv2.imread(saliency_path, cv2.IMREAD_GRAYSCALE)
    saliency = cv2.resize(saliency, img_size)
    saliency = saliency / 255.0
    return saliency
```

```python
num = []
```

```python
# Load and preprocess dataset
cwd = os.getcwd()
# data_files = sorted(os.listdir(cwd))
data_1 = os.path.join(cwd, 'data-1') 

images = []
saliency_maps = []

image_file = os.path.join(data_1, 'imgs')
image_paths = sorted(os.listdir(image_file))
saliency_file = os.path.join(data_1, 'img-saliency')
saliency_paths = sorted(os.listdir(saliency_file))

i = 1
j = 1

for image_path in image_paths:
  image = load_image(os.path.join(image_file, image_path))
  images.append(image)
  for saliency_path in saliency_paths:
    if saliency_path == image_path:
      saliency_map = load_saliency_map(os.path.join(saliency_file, saliency_path))
      saliency_maps.append(saliency_map)
      print(f'{i} {os.path.basename(image_path)} is loaded!')
      i = i + 1
      j = j + 1
      break


num.append(j-1)
```

### Train the CNN model

```python
from keras.applications import VGG16
from keras.models import Sequential
from keras.layers import Flatten, Dense, UpSampling2D, Reshape, Conv2D
from sklearn.model_selection import train_test_split
from skimage.transform import resize
from skimage.color import gray2rgb
import numpy as np

# Set image size
img_size = (224, 224)

# Load pre-trained model
base_model = VGG16(weights='imagenet', include_top=False, input_shape=img_size + (3,))

# Freeze layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom layers
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(np.prod(base_model.output_shape[1:]), activation='linear'))
model.add(Reshape(base_model.output_shape[1:]))
model.add(UpSampling2D(size=(img_size[0]//base_model.output_shape[1], img_size[1]//base_model.output_shape[2])))
model.add(Conv2D(1, (3, 3), padding='same', activation='sigmoid'))

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Upsample saliency maps to match the output shape of the last layer of the VGG16 base model
upsampled_saliency_maps = np.zeros((len(saliency_maps),) + img_size + (1,))
for i, saliency_map in enumerate(saliency_maps):
    upsampled_saliency_maps[i] = np.expand_dims(resize(saliency_map, img_size), axis=-1)

# Split dataset into training, validation and testing sets
train_val_images, test_images, train_val_saliency_maps, test_saliency_maps = train_test_split(images, upsampled_saliency_maps, test_size=0.2, random_state=42)
train_images, val_images, train_saliency_maps, val_saliency_maps = train_test_split(train_val_images, train_val_saliency_maps, test_size=0.2, random_state=42)
```

```python
# Define the filepath and filename for the saved model
filepath = '/content/drive/MyDrive/FYP Year 3/Coding/models/m1.h5'
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

# Define the ModelCheckpoint callback to save the best model based on validation loss
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', save_best_only=True, mode='min', verbose=1)

# Define the EarlyStopping callback to stop training if validation loss does not improve for 3 epochs
earlystop = EarlyStopping(monitor='val_loss', patience=5)

# Train the model with the callbacks
model.fit(train_images, train_saliency_maps, epochs=20, batch_size=32, validation_data=(val_images, val_saliency_maps), callbacks=[checkpoint, earlystop])
```

```python
# Load the saved model for prediction
from keras.models import load_model
saved_model = load_model(filepath)
```

```python
# Get predictions on test set
test_preds = saved_model.predict(test_images)
```

```python
# Set threshold value for binarization
threshold = 0.5

# Binarize predictions based on threshold
binarized_preds = (test_preds > threshold).astype(int)
test_saliency_maps = test_saliency_maps.astype(int)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Calculate accuracy, precision, recall, and f1-score
accuracy = np.mean(binarized_preds == test_saliency_maps)
precision = precision_score(test_saliency_maps.flatten(), binarized_preds.flatten())
recall = recall_score(test_saliency_maps.flatten(), binarized_preds.flatten())
f1 = f1_score(test_saliency_maps.flatten(), binarized_preds.flatten())

print("Test set accuracy: {:.2%}".format(accuracy))
print("Test set precision: {:.2%}".format(precision))
print("Test set recall: {:.2%}".format(recall))
print("Test set f1-score: {:.2%}".format(f1))
```

```python
def binarize_saliency_maps(saliency_maps, threshold):
    saliency_maps[saliency_maps < threshold] = 0
    saliency_maps[saliency_maps >= threshold] = 1
    return saliency_maps.astype('uint8')

binarized_preds = binarize_saliency_maps(test_preds_norm, threshold)
```

```python
test_saliency_maps_int = test_saliency_maps.astype(int)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Calculate accuracy, precision, recall, and f1-score
accuracy = np.mean(binarized_preds == test_saliency_maps)
precision = precision_score(test_saliency_maps_int.flatten(), binarized_preds.flatten())
recall = recall_score(test_saliency_maps_int.flatten(), binarized_preds.flatten())
f1 = f1_score(test_saliency_maps_int.flatten(), binarized_preds.flatten())

print("Test set accuracy: {:.2%}".format(accuracy))
print("Test set precision: {:.2%}".format(precision))
print("Test set recall: {:.2%}".format(recall))
print("Test set f1-score: {:.2%}".format(f1))
```
