## Preprocess and Load Data

```python
import cv2
import os
import numpy as np
```

```python
# Define image size
img_size = (224, 224)

# Define function to load and preprocess images
def load_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, img_size)
    img = img / 255.0
    return img

# Define function to load and preprocess ground truth saliency maps
def load_saliency_map(saliency_path):
    saliency = cv2.imread(saliency_path, cv2.IMREAD_GRAYSCALE)
    saliency = cv2.resize(saliency, img_size)
    saliency = saliency / 255.0
    return saliency
```

## Generate Saliency Map

```python
import cv2
import numpy as np

def generate_saliency_maps(img):
    # Convert the image to grayscale
    gray = cv2.cvtColor(img.astype(np.float32), cv2.COLOR_RGB2GRAY)

    # Calculate the spectral residual
    fft = np.fft.fft2(gray)
    log_amplitude = np.log(np.abs(fft) + 1)
    phase = np.angle(fft)
    spectral_residual = np.exp(log_amplitude - cv2.boxFilter(log_amplitude, ddepth=-1, ksize=(3, 3)))

    # Multiply the spectral residual with a Gaussian pyramid of the image
    pyramid_levels = 3
    pyramid = [gray]
    for i in range(pyramid_levels):
        pyramid.append(cv2.pyrDown(pyramid[-1]))
    pyramid = pyramid[::-1]
    saliency_map = np.zeros_like(gray)
    for i in range(pyramid_levels):
        gray = pyramid[i]
        sr = cv2.resize(spectral_residual, (gray.shape[1], gray.shape[0]))
        fm = np.multiply(np.power(np.abs(sr), 2), gray)
        fm = cv2.GaussianBlur(fm, (3, 3), 0)
        saliency_map += cv2.resize(fm, (saliency_map.shape[1], saliency_map.shape[0]))

    # Normalize the saliency map
    saliency_map = (saliency_map - saliency_map.min()) / (saliency_map.max() - saliency_map.min())

    return saliency_map
```

```python
generated_saliency_maps = []
for img in images:
    saliency_map = generate_saliency_maps(img)
    generated_saliency_maps.append(saliency_map)
```

```python
binarised_saliency_maps = []
threshold = 0.5
for saliency_map in generated_saliency_maps:
    binary_map = np.zeros_like(saliency_map)
    binary_map[saliency_map >= threshold] = 1
    binarised_saliency_maps.append(binary_map)
```

```python
saliency_maps_r = np.ravel(saliency_maps)
binarised_saliency_maps_r = np.ravel(binarised_saliency_maps)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_saliency_maps(generated_maps_binary, ground_truth_maps):

    # Calculate the evaluation metrics
    accuracy = accuracy_score(ground_truth_maps, generated_maps_binary)
    precision = precision_score(ground_truth_maps, generated_maps_binary)
    recall = recall_score(ground_truth_maps, generated_maps_binary)
    f1 = f1_score(ground_truth_maps, generated_maps_binary)
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")

    generated_maps_binary = generated_maps_binary.reshape(np.array(generated_saliency_maps).shape)
    ground_truth_maps = ground_truth_maps.reshape(np.array(generated_saliency_maps).shape)
    
    # Find the indices with the highest f1-score
    f1_scores = []
    for i, (gen, gt) in enumerate(zip(generated_maps_binary, ground_truth_maps)):
        gt = np.ravel(gt)
        gen = np.ravel(gen)
        f1_scores.append(f1_score(gt, gen))
    top_indices = sorted(range(len(f1_scores)), key=lambda i: f1_scores[i], reverse=True)[:10]

    for idx in top_indices:
        print(f1_scores[idx])

    
    return top_indices
```

```python
top_indices = evaluate_saliency_maps(binarised_saliency_maps_r, saliency_maps_r)
```

```python
# Plot the images with the highest f1-score
fig, axs = plt.subplots(10, 4, figsize=(25, 25))
for i, idx in enumerate(top_indices):
    axs[i, 0].imshow(images[idx])
    axs[i, 1].imshow(generated_saliency_maps[idx])
    axs[i, 2].imshow(binarised_saliency_maps[idx], cmap = 'gray')
    axs[i, 3].imshow(saliency_maps[idx], cmap = 'gray')
    axs[i, 0].set_title("Original Image " + str(idx))
    axs[i, 1].set_title("Saliency Map")
    axs[i, 2].set_title("Binarised")
    axs[i, 3].set_title("Ground-truth")
    axs[i, 0].axis('off')
    axs[i, 1].axis('off')
    axs[i, 2].axis('off')
    axs[i, 3].axis('off')
fig.tight_layout()
```
