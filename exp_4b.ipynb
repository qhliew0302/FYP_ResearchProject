### Preprocess and Load Data

```python
import cv2
import os
import numpy as np
```

```python
# Define image size
img_size = (64, 64)

# Define function to load and preprocess images
def load_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, img_size)
    img = img / 255.0
    return img

# Define function to load and preprocess ground truth saliency maps
def load_saliency_map(saliency_path):
    saliency = cv2.imread(saliency_path, cv2.IMREAD_GRAYSCALE)
    saliency = cv2.resize(saliency, img_size)
    saliency = saliency / 255.0
    return saliency
```

### Gradient-based

```python
from keras.models import Model
from keras.applications.vgg16 import VGG16
from keras.layers import Conv2D, Conv2DTranspose, Input, MaxPooling2D, UpSampling2D, Lambda
import tensorflow as tf
import keras.backend as K
tf.compat.v1.disable_eager_execution()

def build_gradient_model():
    # define input shape
    input_shape = (64, 64, 3)

    # create input tensor
    input_tensor = Input(shape=input_shape)

    # load VGG16 model pre-trained on ImageNet
    vgg16_model = VGG16(weights='imagenet', input_tensor=input_tensor, include_top=False)

    # Get the output of the last convolutional layer
    last_conv_layer = vgg16_model.get_layer('block5_conv3')

    # Create a new convolutional layer to increase the output shape
    x = Conv2D(512, (3, 3), activation='relu', padding='same', name='new_conv')(last_conv_layer.output)

    # Create a new model with the same inputs as the original model
    inputs = vgg16_model.inputs
    outputs = x
    model = Model(inputs=inputs, outputs=outputs)

    # Define the loss function as the mean of the output tensor
    loss = K.mean(outputs)

    # Compute the gradient of the loss with respect to the input image
    grads = K.gradients(loss, inputs)[0]

    # Normalize the gradients
    grads /= K.sqrt(K.mean(K.square(grads))) + K.epsilon()

    # Define the function to compute the saliency maps
    saliency_fn = K.function(inputs, [grads])

    # Create a new model to output the saliency maps
    saliency_map = K.abs(grads)
    saliency_map = K.max(saliency_map, axis=3)
    saliency_map = K.expand_dims(saliency_map)
    saliency_map = K.resize_images(saliency_map, input_shape[0], input_shape[1], data_format='channels_last')

    # Add a new Lambda layer to resize the output to the desired shape
    saliency_map = Lambda(lambda x: tf.image.resize(x, [64, 64], method='bilinear'))(saliency_map)

    saliency_model = Model(inputs=inputs, outputs=saliency_map)

    return saliency_model, saliency_fn
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Split the data into train and test sets
from sklearn.model_selection import train_test_split
# Split dataset into training, validation and testing sets
train_val_images, test_images, train_val_saliency_maps, test_saliency_maps = train_test_split(images, saliency_maps, test_size=0.2, random_state=42)
train_images, val_images, train_saliency_maps, val_saliency_maps = train_test_split(train_val_images, train_val_saliency_maps, test_size=0.2, random_state=42)
```

```python
print(f"Number of test images: {len(test_images)}")
print(f"Number of test saliency maps: {len(test_saliency_maps)}")
```

```python
# Build and compile the model
model, saliency_fn = build_gradient_model()
# model.compile(loss='binary_crossentropy', optimizer='adam')
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Define the filepath and filename for the saved model
filepath = '/content/drive/MyDrive/FYP Year 3/Coding/models/m4.h5'
```

```python
model.summary()
```

```python
from skimage.transform import resize
input_shape = (64, 64, 3)
train_saliency_maps_resized = np.array([resize(img, input_shape[:2]) for img in train_saliency_maps])
train_saliency_maps_resized = np.expand_dims(train_saliency_maps_resized, axis=-1)
val_saliency_maps_resized = np.array([resize(img, input_shape[:2]) for img in val_saliency_maps])
val_saliency_maps_resized = np.expand_dims(val_saliency_maps_resized, axis=-1)
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

# Define the ModelCheckpoint callback to save the best model based on validation loss
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', save_best_only=True, mode='min', verbose=1)

# Define the EarlyStopping callback to stop training if validation loss does not improve for 3 epochs
earlystop = EarlyStopping(monitor='val_loss', patience=5)

# Train the model with the callbacks
model.fit(train_images, train_saliency_maps_resized, epochs=30, batch_size=32, validation_data=(val_images, val_saliency_maps_resized), callbacks=[checkpoint, earlystop])
```

```python
# Load the saved model for prediction
from keras.models import load_model
saved_model = load_model(filepath)
```

```python
# Get predictions on test set
pred_saliency_maps = saved_model.predict(test_images)
```

```python
print(f"Shape of predicted saliency maps: {pred_saliency_maps.shape}")
```

```python
pred_saliency_maps_resized = []
for pred_map in pred_saliency_maps:
    resized_map = cv2.resize(pred_map, img_size, interpolation=cv2.INTER_LINEAR)
    # Normalize the saliency map to have values between 0 and 1
    resized_map = (resized_map - resized_map.min()) / (resized_map.max() - resized_map.min())
    pred_saliency_maps_resized.append(resized_map)

pred_saliency_maps_resized = np.array(pred_saliency_maps_resized)
```

```python
import numpy as np
test_saliency_maps_binary = np.where(test_saliency_maps > 0, 1, 0)
pred_saliency_maps_binary = np.where(pred_saliency_maps_resized > 0.2, 1, 0)
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(test_saliency_maps_binary.ravel(), pred_saliency_maps_binary.ravel())
precision = precision_score(test_saliency_maps_binary.ravel(), pred_saliency_maps_binary.ravel())
recall = recall_score(test_saliency_maps_binary.ravel(), pred_saliency_maps_binary.ravel())
f1 = f1_score(test_saliency_maps_binary.ravel(), pred_saliency_maps_binary.ravel())

print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1-score: {f1}")
```
