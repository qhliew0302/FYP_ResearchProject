## Preprocess and Load Data

```python
import cv2
import os
import numpy as np
```

```python
# Define image size
img_size = (224, 224)

# Define function to load and preprocess images
def load_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, img_size)
    img = img / 255.0
    return img

# Define function to load and preprocess ground truth saliency maps
def load_saliency_map(saliency_path):
    saliency = cv2.imread(saliency_path, cv2.IMREAD_GRAYSCALE)
    saliency = cv2.resize(saliency, img_size)
    saliency = saliency / 255.0
    return saliency
```

## Generate Saliency Maps

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def generate_saliency_maps(img):
    # Convert the image to grayscale
    gray = cv2.cvtColor(img.astype(np.float32), cv2.COLOR_RGB2GRAY)

    # Define the Gaussian pyramid levels
    pyramid_levels = 4

    # Calculate the Gaussian pyramid
    pyramid = [gray]
    for i in range(pyramid_levels):
        pyramid.append(cv2.pyrDown(pyramid[-1]))

    # Calculate the feature maps
    intensity_map = gray
    img = (img * 255).astype(np.uint8)
    color_map = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)[:, :, 1:]
    grad_x = np.array([cv2.Sobel(cv2.resize(x.astype(np.uint8), pyramid[0].shape[::-1]), cv2.CV_64F, 1, 0, ksize=3) for x in pyramid], dtype=np.float32)
    grad_y = np.array([cv2.Sobel(cv2.resize(x.astype(np.uint8), pyramid[0].shape[::-1]), cv2.CV_64F, 0, 1, ksize=3) for x in pyramid], dtype=np.float32)

    grad_xy = np.stack([grad_x, grad_y], axis=1) # stack gradient images for x and y directions

    orientation_map = cv2.phase(grad_xy[:,0], grad_xy[:,1]) # calculate the orientation map using cv2.phase()
                
    # Calculate the center-surround differences
    intensity_cs = []
    color_cs = []
    orientation_cs = []
    for i in range(pyramid_levels):
        inten_diff = cv2.GaussianBlur(intensity_map, (0, 0), sigmaX=2**(i+2)) - cv2.resize(cv2.GaussianBlur(intensity_map, (0, 0), sigmaX=2**(i+1)), 
                                                                                                (intensity_map.shape[1], intensity_map.shape[0]))
        intensity_cs.append(cv2.resize(inten_diff, (img.shape[1], img.shape[0])))
        
        color_diff = cv2.GaussianBlur(color_map, (0, 0), sigmaX=2**(i+2), sigmaY=2**(i+2)) - cv2.resize(cv2.GaussianBlur(color_map, (0, 0), sigmaX=2**(i+1), sigmaY=2**(i+1)), (color_map.shape[1], color_map.shape[0]))

        color_cs.append(cv2.resize(color_diff, (img.shape[1], img.shape[0]))[:, :, 1:])
        
        orient_diff = cv2.GaussianBlur(orientation_map[i], (0, 0), sigmaX=2**(i+2)) - cv2.resize(cv2.GaussianBlur(orientation_map[i], (0, 0), sigmaX=2**(i+1)), 
                                                                                                    (orientation_map[i].shape[1], orientation_map[i].shape[0]))
        orientation_cs.append(cv2.resize(orient_diff, (img.shape[1], img.shape[0])))

    intensity_cs = [np.reshape(x, (x.shape[0], x.shape[1], 1)) for x in intensity_cs]
    color_cs = [np.reshape(x, (x.shape[0], x.shape[1], 1)) for x in color_cs]

    # Combine the feature maps to form the saliency map
    saliency_map = np.sum([cv2.GaussianBlur(intensity_cs[i]**2, (0, 0), sigmaX=2) +
                          cv2.GaussianBlur(color_cs[i]**2, (0, 0), sigmaX=2) +
                          cv2.GaussianBlur(orientation_cs[i]**2, (0, 0), sigmaX=2)
                          for i in range(pyramid_levels)], axis=0)

    # Normalize the saliency map
    saliency_map = (saliency_map - saliency_map.min()) / (saliency_map.max() - saliency_map.min())

    return saliency_map
```

```python
generated_saliency_maps = []
for img in images:
    saliency_map = generate_saliency_maps(img)
    generated_saliency_maps.append(saliency_map)
```

```python
binarised_saliency_maps = []
threshold = 0.5
for saliency_map in generated_saliency_maps:
    binary_map = np.zeros_like(saliency_map)
    binary_map[saliency_map >= threshold] = 1
    binarised_saliency_maps.append(binary_map)
```

```python
saliency_maps_r = np.ravel(saliency_maps)
binarised_saliency_maps_r = np.ravel(binarised_saliency_maps)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_saliency_maps(generated_maps_binary, ground_truth_maps):

    # Calculate the evaluation metrics
    accuracy = accuracy_score(ground_truth_maps, generated_maps_binary)
    precision = precision_score(ground_truth_maps, generated_maps_binary)
    recall = recall_score(ground_truth_maps, generated_maps_binary)
    f1 = f1_score(ground_truth_maps, generated_maps_binary)
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")

    generated_maps_binary = generated_maps_binary.reshape(np.array(generated_saliency_maps).shape)
    ground_truth_maps = ground_truth_maps.reshape(np.array(generated_saliency_maps).shape)
    
    # Find the indices with the highest f1-score
    f1_scores = []
    for i, (gen, gt) in enumerate(zip(generated_maps_binary, ground_truth_maps)):
        gt = np.ravel(gt)
        gen = np.ravel(gen)
        f1_scores.append(f1_score(gt, gen))
    top_indices = sorted(range(len(f1_scores)), key=lambda i: f1_scores[i], reverse=True)[:8]

    for idx in top_indices:
        print(f1_scores[idx])
    return top_indices
```

```python
top_indices = evaluate_saliency_maps(binarised_saliency_maps_r, saliency_maps_r)
```

```python
# Plot the images with the highest f1-score
fig, axs = plt.subplots(8, 4, figsize=(25, 25))
for i, idx in enumerate(top_indices):
    axs[i, 0].imshow(images[idx])
    axs[i, 1].imshow(generated_saliency_maps[idx])
    axs[i, 2].imshow(binarised_saliency_maps[idx], cmap = 'gray')
    axs[i, 3].imshow(saliency_maps[idx], cmap = 'gray')
    axs[i, 0].set_title("Original Image " + str(idx))
    axs[i, 1].set_title("Saliency Map")
    axs[i, 2].set_title("Binarised")
    axs[i, 3].set_title("Ground-truth")
    axs[i, 0].axis('off')
    axs[i, 1].axis('off')
    axs[i, 2].axis('off')
    axs[i, 3].axis('off')
fig.tight_layout()
```
